```python
import cv2
import matplotlib.pyplot as plt
import lane_detection as lp
```

```python
image = cv2.imread("./data/img/01. solidWhiteCurve.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
color_selected = lp.hsl_color_selection(image)
plt.imshow(color_selected)
```

```python
grayscale = cv2.cvtColor(color_selected, cv2.COLOR_RGB2GRAY)
plt.imshow(grayscale, cmap="gray")
```

```python
kernel_size = 5
blur = cv2.GaussianBlur(grayscale, (kernel_size, kernel_size), 0)
plt.imshow(blur, cmap="gray")
```

```python
low_t = 50
high_t = 150
edges = cv2.Canny(grayscale, low_t, high_t)
plt.imshow(edges, cmap="gray")
```

```python
region = lp.region_selection(edges)
plt.imshow(region, cmap="gray")
```

```python
hough = lp.hough_transform(region)
result = lp.draw_lane_lines(image, lp.lane_lines(image, hough))
plt.imshow(result)
```
